### Atividade - Ajustando o titanic!

* <font size = '4'> Criar um novo dataframe filtrando somente as colunas Survived, Pclass, Age e Fare do dataset;

```python
dataframe_atividade = dataframe[['Survived','Pclass','Age','Fare']].copy()
dataframe_atividade.head()
```

* <font size = '4'> Criar uma função lambda para alterar a formatação da coluna Survived para ter duas casas decimais;

```python
dataframe_atividade['Survived'] = dataframe_atividade['Survived'].apply(lambda x : "%.2f" % float(x))
```

```python
dataframe_atividade.head()
```

* <font size = '4'> Criar uma função lambda para alterar a formatação da coluna Pclass para ter três casas decimais;

```python
dataframe_atividade['Pclass'] = dataframe_atividade['Pclass'].apply(lambda x : f"{ float(x):.3f}")
```

```python
dataframe_atividade.head()
```

* <font size = '4'>  Criar uma função lamba para alterar a formatação de todo o dataframe para usar somente uma casa decimal;

```python
dataframe_atividade = dataframe_atividade.applymap(lambda x: '%.1f' % float(x))
```

```python
dataframe_atividade
```
