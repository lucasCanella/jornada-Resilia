### Atividade - Ticket titanic.

* <font size = '4'> Desenvolver um código que faça o calculo das três medidas estatísticas utilizando dataframe Titanic;

```python
medidas = pd.Series(['Média', 'Moda', 'Mediana'])
medidas_survived = pd.Series([df_titanic['Survived'].mean(), df_titanic['Survived'].mode().iloc[0], df_titanic['Survived'].median()])
medidas_pclass = pd.Series([df_titanic['Pclass'].mean(), df_titanic['Pclass'].mode().iloc[0], df_titanic['Pclass'].median()])
medidas_age = pd.Series([df_titanic['Age'].mean(), df_titanic['Age'].mode().iloc[0], df_titanic['Age'].median()])
medidas_fare = pd.Series([df_titanic['Fare'].mean(), df_titanic['Fare'].mode().iloc[0], df_titanic['Fare'].median()])
```

```python
estatisticas = pd.concat([medidas.rename('Medidas'), medidas_survived.rename('Survived'), medidas_pclass.rename('Pclass'),medidas_age.rename('Age'),medidas_fare.rename('Fare')], axis =1)
estatisticas.set_index('Medidas')
```

* <font size = '4'> Verificar quais são os títulos das pessoas que embarcaram na terceira classe;

```python
def extrai_titulo(df):
  df['Titulo'] = df['Name'].apply(lambda x: x.split(',')[0].split('.')[0].strip().lower())

extrai_titulo(df_titanic)
```
